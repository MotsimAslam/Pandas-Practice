### Animated Libraries

#### We will use plotly to make animated plots.

```python
# 01- Import Library
import plotly.express as px
```

```python
# 02- Import Data
df = px.data.gapminder()
df.head()
```

```python
fig1=px.scatter(df, x='gdpPercap', y='lifeExp',
            animation_frame='year', animation_group='continent', size='pop', 
            color='country', hover_name='country',log_x=True, size_max=55, 
            range_x=[100,100000], range_y=[25,90])
fig1.show()
```

```python
# export the plot in gif or video
fig1.write_html('./output/animated_ploty.gif')
```

```python
# sunbrust Plot
fig=px.sunburst(df,path=['continent','country', 'year'],
                values='pop',color='continent')
fig.show()
```
