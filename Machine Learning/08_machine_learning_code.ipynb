## Model Notebook for Coding in SK-LEARN


```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
# Load the data set of 'tips'
df=sns.load_dataset('tips')
df.head()
```

```python
# Preprocessing the data
# scaler=StandardScaler()
# df[['total_bill', 'tip']]=scaler.fit_transform(df[['total_bill', 'tip']])
```

```python
# Seprate the feature 'X' and Target 'y'
X=df[['total_bill']]
y=df['tip']
```

```python
# train test and split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)
```

```python
# call the model
model= LinearRegression()
```

```python
# train the model
model.fit(X_train, y_train)
```

```python
# predit the model
prediction=model.predict(X_test)
prediction
```

```python
# evalate the model
from sklearn.metrics import mean_squared_error, r2_score
print('MSE', mean_squared_error(y_test, prediction))
print('rse', r2_score(y_test, prediction))
```

```python
# Save the model
import pickle
pickle.dump(model, open('./saved_models/model01.pkl', 'wb'))
```
