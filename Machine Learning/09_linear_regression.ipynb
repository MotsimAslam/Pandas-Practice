# Linear Regression
Linear Regression is a type of regression algorithm that fits a linear function to a set of data points. It is a form of supervised learning. We will be using it in our project.
There are tow tupes of linear regression in machine learning.
 * Linear Regression
 * Multiple Linear Regression

```python
# Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Import the data from sns tips
df=sns.load_dataset('tips')
df.head()
```

```python
# split the data into X and y
X= df[['total_bill']]
y= df['tip']
```

```python
# split the date into train and test
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2)
```

```python
# call the model
model= LinearRegression()
```

```python
# train the model
model.fit(X_train, y_train)
```

```python
# take out model intercept and slope, make an equation
print(model.intercept_)
print(model.coef_)
print('y= ', model.intercept_, '+', model.coef_, '*', 'X')
```

```python
# predit the model
y_pred=model.predict(X_test)
```

```python
# Evaluate the model
print('MSE=' ,mean_squared_error(y_test, y_pred))
print('R2=' ,r2_score(y_test, y_pred))
print('RMSE=' ,np.sqrt(mean_squared_error(y_test, y_pred))) 
```

```python
# plot the model and data
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red')
plt.show()
```
