```python
# importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Types of plot?

```python
# importing dataset
df=sns.load_dataset('titanic')
```

```python
# check the data set
df.info()
```

```python
# scatter ploting in seaborn
sns.scatterplot(x='age',y='fare',data=df,hue='sex')
```

```python
# scatter plot
# scatter plto size of figure size
plt.figure(figsize=(8,4))
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
```

```python
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(8,4)) # a4 page size 16:9

# change the scatter point colors
palette={'male':'green', 'female':'red'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
#Labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

```python
# scatter plot
sns.set_theme(style='darkgrid', font_scale=1.4 , font='cambria')
# scatter plot size of figure size
plt.figure(figsize=(10,4)) # a4 page size 16:9

# change the scatter point colors
palette={'male':'green', 'female':'red'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
#Labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.savefig('./output/pictures/Scatter Plot of Age vs Fare.png', dpi=300)
```

### Interpertation:
1. We have age of pessanger at x-axix and fare at y-axix.
2. The scatter plot show the relationship between age and fare of titanic passenger.
3. We further grouped the pessanger by their gender (such as male is represented as green points) and female is represented as (red points).
4. The two male passenger give the highest fare, might be they are trevelling in first class.
5. The scatter plot also show that there was a male passenger in titanic age 80 years.

```python
# Box plot
sns.set_theme(style='darkgrid')
# Box plot size of figure size
plt.figure(figsize=(12,4)) # a4 page size 16:9

# change the scatter point colors
palette={'First':'green', 'Second':'red', 'Third':'blue'}
sns.boxplot(data=df,x='sex',y='age',hue='class',palette=palette)
#Labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot of Age vs Fare')
##plt.savefig('./output/pictures/Scatter Plot of Age vs Fare.png', dpi=300)

# change legend position
plt.legend(loc= 'upper left', title='Class Name')
```

```python
# Boxen plot
sns.set_theme(style='darkgrid')
# Box plot size of figure size
plt.figure(figsize=(16,9)) # a4 page size 16:9

# change the scatter point colors
palette={'First':'green', 'Second':'red', 'Third':'blue'}
sns.boxenplot(data=df,x='sex',y='age',hue='class',palette=palette)
#Labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot of Age vs Fare')
##plt.savefig('./output/pictures/Scatter Plot of Age vs Fare.png', dpi=300)

# change legend position
plt.legend(loc= 'upper left', title='Class Name')
```
