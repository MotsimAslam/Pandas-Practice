##Visulization by using Pandas

```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
#load the data set
df=sns.load_dataset('titanic')
```

```python
# 01- Scatter plot
df.plot(x='age',y='fare',kind='scatter', color='red')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.ylabel('Age')
plt.xlabel('Sex')
```

```python
# check the data head
df.head()
```

```python
# 02- Bar plot by using Seaborn
sns.barplot(x='fare',y='sex',data=df, hue='class')
```
