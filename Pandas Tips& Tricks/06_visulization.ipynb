##Visulization by using Pandas

```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
#load the data set
df=sns.load_dataset('titanic')
```

```python
# 01- Scatter plot
df.plot(x='age',y='fare',kind='scatter', color='red')
```

```python
# 02- Bar plot
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.ylabel('Age')
plt.xlabel('Sex')
```

```python
# check the data head
df.head()
```

```python
# 02- Bar plot by using Seaborn
sns.barplot(x='sex',y='fare',data=df, hue='class')
```

```python
#pip install plotly
#pip install nbformat>=4.2.0
import plotly.express as px
```

```python
# make scatter plot via plotly
fig= px.scatter(df, x='age', y='fare', color='class')
fig.show()
```

## Using all Plots Step by Step

```python
# import pandas and matplotlib.pyplot libraries
import pandas as pd
import matplotlib.pyplot as plt

# create a dictionary with sample data
data = {'Products': ['Product A', 'Product B', 'Product C'],
    'Sales': [100, 150, 80]}

# create a pandas dataframe from the dictionary
df = pd.DataFrame(data)

# plot the dataframe as a bar chart
df.plot(x='Products', y='Sales', kind='bar', legend=False)

# set the y-axis label
plt.ylabel('Sales')

# set the title of the plot
plt.title('Sales of different products')

# save the plot to a file
plt.savefig('bar_chart.png', dpi=300, bbox_inches='tight')

# display the plot
plt.show()
```
